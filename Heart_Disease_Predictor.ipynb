```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("heart_disease_data.csv")
df
```

```python
#separating features and target variables
X = df.drop(columns = ["target"])
y = df["target"]
```

```python
#defining categorical and numerical terms
cat_data = ["sex" , "cp", "restecg" , "slope", "thal"]
num_data = ["age", "trestbps" , "chol" , "thalach" , "oldpeak" ,"ca"]
```

```python
numerical_transformer = MinMaxScaler()
categorical_transformer = OneHotEncoder(drop='first', sparse=False)
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, num_data),
        ('cat', categorical_transformer, cat_data)
    ])
```

```python
X_processed = preprocessor.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)
```

```python
#now on applying the regression model we get:

model = LogisticRegression(max_iter = 1000 , random_state = 42)

#training
model.fit(X_train, y_train)

#predictions on the testing set
y_pred = model.predict(X_test)

#calculating accuracy
accuracy = accuracy_score(y_test , y_pred)
print("Accuracy : ", accuracy)
```

```python
X_test_indexed = pd.DataFrame(X_test, index=y_test.index)

# Add a new column 'predicted_target' to the DataFrame
X_test_indexed['predicted_target'] = y_pred
pred_df = df.merge(X_test_indexed, left_index=True, right_index=True)
print(pred_df)
```

```python
# Define the filename for the CSV file
output_filename = "updated_dataframe.csv"
pred_df.to_csv(output_filename, index=False)
print("DataFrame has been saved as:", output_filename)
```
